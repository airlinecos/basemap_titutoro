```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
map = Basemap(llcrnrlon=13.6529,llcrnrlat=40.5270,urcrnrlon=14.7029,urcrnrlat=40.9270, resolution = 'h', projection='cyl')

fig = plt.figure(figsize=(14,14))
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='#ddaa66',lake_color='aqua')
map.drawcoastlines()

x,y = map(14.199255, 40.827582)
map.plot(x,y,marker='D',color='r')
plt.show()
```

```python
map = Basemap(resolution='l', 
              satellite_height=3600000.,
              projection='nsper', 
              lat_0 = 30., lon_0 = -10.,
              llcrnrx=500000.,llcrnry=500000.,urcrnrx=2700000.,urcrnry=2700000.
             )

fig = plt.figure(figsize=(14,14))
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='#ddaa66',lake_color='aqua')

map.drawcoastlines()

x, y = map(8.942228, 46.008869)
map.plot(x, y, marker='D', color='k')

x2, y2 = map(7.1251, 43.5804)
map.plot(x2, y2, marker='D', color='b')

x3, y3 = map(14.2681, 40.8518)
map.plot(x3, y3, marker='D', color='r')

plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from osgeo import gdal
import numpy

fig = plt.figure(figsize=(12,12))
map = Basemap(llcrnrlon=5, llcrnrlat=35., urcrnrlon=20, urcrnrlat=50,
              projection='merc', resolution='h')#lat_1=30., lat_2=60., lat_0=34.83158, lon_0=-98.)

ds = gdal.Open("sample_files/wrf.tiff")
lons = ds.GetRasterBand(4).ReadAsArray()
lats = ds.GetRasterBand(5).ReadAsArray()
u10 = ds.GetRasterBand(1).ReadAsArray()
v10 = ds.GetRasterBand(2).ReadAsArray()
speed = numpy.sqrt(u10*u10 + v10*v10)
xx, yy = map.makegrid(v10.shape[1], v10.shape[0], returnxy=True)[2:4]

map.streamplot(xx, yy, u10, v10, color=speed, cmap=plt.cm.autumn, linewidth=0.5*speed)
map.bluemarble()
map.drawcoastlines(color = '0.15')

plt.show()
```
